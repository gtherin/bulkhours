# Evaluation of the content of an image with machine learning üêç

**Using a ResNet50 network to identify the main subject of a photo:**



```
%%evaluation_cell_id -i aimg

def get_predictions():

    # Get and format image
    img = image.load_img(bulkhours.get_data("chose1.jpg"), target_size=(224, 224))

    # Display image
    IPython.display.display(img)

    # Format the image
    rimg = np.expand_dims(image.img_to_array(img), axis=0)

    # Use ResNet50 from tensorflow
    model = bulkhours.hpc.get_tf_model("ResNet50");

    # Get prediction
    preds = model.predict(preprocess_input(rimg))

    # Print classification
    return pd.DataFrame(decode_predictions(preds, top=3)[0], columns=["idx", "Evaluation", "Probabilit√©"]).drop(columns=["idx"])

get_predictions()
```
