# Initialization of the environnement

First, you need to install the package from github and import it.
Like in the previous tutorial, we should enter an email.

```
# Install the bulkhours package, redirecting stdout and stderr to a null output
!rm -rf bulkhours && git clone https://github.com/guydegnol/bulkhours.git > /dev/null 2>&1

# Import the bulkhours package
import bulkhours

# Like in the previous tutorial, we should enter an email. On top of that, we will add several more info
bulkhours.init_env(
    email="yo.da@jedi.com",             # Email of the teacher
    subject="s",                       # Recommended to be unique for an entire course
    notebook_id="course_edition",      # Recommended to be unique per notebook (id of the current notenook)
    is_admin=True,                     # Switch to admin mode (Ok for the demo mode)
    database="data/cache/starwars.json" # Database file
    )

# Generate header links
bulkhours.generate_header_links("examples/2_Course_Edition.ipynb")
```
