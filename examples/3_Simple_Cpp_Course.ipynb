# Somme de vecteurs sur carte graphique

**G√©n√©rer la somme de 2 vecteurs avec CUDA:**


```
%%evaluation_cell_id -i cuda -w sc -l G√©n√©rer la somme de 2 vecteurs avec CUDA
%%compile_and_exec -c nvcc
#include <stdio.h>

// Create sum function
__global__ void sum_vec_gpu(int *a, int *b, int *c) {
    int i = threadIdx.x + blockDim.x * blockIdx.x;
    c[i] = a[i] + b[i];
}

int main(int argc, char *argv[]) {
    // Define and allocate Host memory for a vector
    int a[6] = {4, 4, 4, 4, 4, 4};
    int b[6] = {0, 1, 2, 3, 4, 5};
    int * c; c = (int*)malloc(6 * sizeof(int));

    // Allocate Device memory for 2 vectors
    int *d_ac; cudaMalloc((void **) &d_ac, 6 * sizeof(int));
    int *d_b; cudaMalloc((void **) &d_b, 6 * sizeof(int));

    // Copy data from Host to Device
    cudaMemcpy(d_ac, a, 6 * sizeof(int), cudaMemcpyHostToDevice);
    cudaMemcpy(d_b, b, 6 * sizeof(int), cudaMemcpyHostToDevice);

    // Use the same variable for a and c on Device: d_ac
    // Kernel invocation with several Threads
    sum_vec_gpu<<<512, 512>>>(d_ac, d_b, d_ac);

    // Copy data from Device to Host (c in enough)
    cudaMemcpy(c, d_ac, 6 * sizeof(int), cudaMemcpyDeviceToHost);

    // Display content of the output vector
    for (int i = 0; i < 6; i++) printf("%d,", c[i]);
}
```
